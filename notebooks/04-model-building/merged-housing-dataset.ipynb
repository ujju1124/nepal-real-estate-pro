```python
import numpy as np   
import pandas as pd
```

```python
df=pd.read_csv('lalpurja_house_v2_features_ready.csv')
# ─────────────────────────────────────────
# STEP 1 — LOAD ORIGINAL HOUSING DATASET
# This is the cleaned housing dataset from
# earlier in the project
# ─────────────────────────────────────────
df_orig = pd.read_csv('housing_features_ready_after_feature_engineering.csv')
```

```python
df_new=df.copy()
```

```python


print(f"=== Original Housing Dataset ===")
print(f"Shape: {df_orig.shape}")
print(f"Columns: {list(df_orig.columns)}")

print(f"\n=== Lalpurja Housing v2 ===")
print(f"Shape: {df_new.shape}")
print(f"Columns: {list(df_new.columns)}")

# Find column differences
orig_cols = set(df_orig.columns)
lph_cols  = set(df_new.columns)

print(f"\n=== In Original but NOT in Lalpurja ===")
print(sorted(orig_cols - lph_cols))

print(f"\n=== In Lalpurja but NOT in Original ===")
print(sorted(lph_cols - orig_cols))

print(f"\n=== Common columns ===")
print(sorted(orig_cols & lph_cols))
```
