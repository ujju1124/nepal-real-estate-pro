```python
# ─────────────────────────────────────────
# STEP 3 — PARSE total_price_raw → total_price (NPR)
# Format examples:
#   "3 Crore 50 Lakh" → 3×10,000,000 + 50×100,000 = 35,000,000
#   "4 Crore"         → 4×10,000,000 = 40,000,000
#   "75 Lakh"         → 75×100,000 = 7,500,000
# Why: ML model needs numeric values not strings
# ─────────────────────────────────────────

def parse_nepali_price(val):
    if pd.isna(val):
        return np.nan
    
    val = str(val).strip().lower()
    total = 0
    
    # Extract crore value if present
    # Example: "3 crore 50 lakh" → crore=3
    crore_match = re.search(r'(\d+\.?\d*)\s*crore', val)
    if crore_match:
        total += float(crore_match.group(1)) * 10_000_000
    
    # Extract lakh value if present
    # Example: "3 crore 50 lakh" → lakh=50
    lakh_match = re.search(r'(\d+\.?\d*)\s*lakh', val)
    if lakh_match:
        total += float(lakh_match.group(1)) * 100_000
    
    return total if total > 0 else np.nan

import re

df_lph['total_price'] = df_lph['total_price_raw'].apply(parse_nepali_price)
```

```python
# ─────────────────────────────────────────
# STEP 4 — PARSE property_area → land_size_aana
# Keeping original property_area column for safety
# Will drop at the end after full verification
# ─────────────────────────────────────────

def parse_property_area(val):
    if pd.isna(val):
        return np.nan

    val = str(val).strip().lower()

    # If no number found → garbage value like "Lalitpur, Balkumari"
    num_match = re.search(r'(\d+\.?\d*)', val)
    if not num_match:
        return np.nan
    num = float(num_match.group(1))

    if 'aana' in val or 'anna' in val:
        # Standard unit — return directly
        return num
    elif 'ropani' in val:
        # 1 ropani = 16 aana
        # Example: "1 Ropani" → 16, "3.5 Ropani" → 56
        return num * 16
    elif 'paisa' in val:
        # 4 paisa = 1 aana
        # Example: "3 Paisa" → 0.75 aana
        return round(num / 4, 3)
    elif 'daam' in val:
        # 16 daam = 1 aana
        # Example: "4.06 Daam" → 0.25 aana
        return round(num / 16, 3)
    elif 'sq. meter' in val or 'sq meter' in val:
        # 1 aana = 31.8 sqm
        # Example: "291.55 Sq. Meter" → 9.17 aana
        return round(num / 31.8, 2)
    elif 'sq' in val or 'feet' in val or 'ft' in val:
        # 1 aana = 342.25 sqft
        # Example: "2000 Sq. Ft." → 5.84 aana
        return round(num / 342.25, 2)

    return np.nan

# Create new column — keep original property_area intact
df_lph['land_size_aana'] = df_lph['property_area'].apply(parse_property_area)

# Verify
print("=== land_size_aana stats ===")
print(df_lph['land_size_aana'].describe())
print(f"\nNulls: {df_lph['land_size_aana'].isnull().sum()}")
print(f"Below 1 aana:  {(df_lph['land_size_aana'] < 1).sum()}")
print(f"Above 50 aana: {(df_lph['land_size_aana'] > 50).sum()}")
print(f"\nSample values:")
print(df_lph['land_size_aana'].value_counts().head(10))
```

```python
# ─────────────────────────────────────────
# STEP 5 — PARSE built_up_area → built_up_sqft
# Keeping original built_up_area intact
# Units found:
#   square-feet → keep as is
#   aana → most are genuine aana BUT
#          large values like 1026.75, 855.625 are
#          actually sqft with wrong label — detect by size
#          Real aana built up: 1–20 aana max
#          Fake aana (actually sqft): values > 50
#   square-meter → multiply by 10.764
#   paisa → 1 aana = 4 paisa → paisa/4 × 342.25
#   plain number with no unit → assume sqft
# ─────────────────────────────────────────

def parse_built_up_area(val):
    if pd.isna(val):
        return np.nan

    val = str(val).strip().lower()

    # Extract numeric value
    num_match = re.search(r'(\d+\.?\d*)', val)
    if not num_match:
        return np.nan
    num = float(num_match.group(1))

    if 'square-feet' in val or 'square feet' in val:
        # Standard sqft — return directly
        return num

    elif 'sq' in val and 'meter' in val:
        # Square meter → convert to sqft
        # 1 sqm = 10.764 sqft
        # Example: "139.09 square-meter" → 1497.23 sqft
        return round(num * 10.764, 2)

    elif 'aana' in val or 'anna' in val:
        # Two cases:
        # Real aana: small values like "3 aana", "4.5 aana" → convert to sqft
        # Fake aana (actually sqft): large values like "1026.75 aana" → keep as sqft
        # Threshold: if value > 50 it's almost certainly sqft mislabeled as aana
        if num > 50:
            # Mislabeled — treat as sqft directly
            # Example: "1026.75 aana" → 1026.75 sqft
            return num
        else:
            # Genuine aana — convert to sqft
            # Example: "3 aana" → 3 × 342.25 = 1026.75 sqft
            return round(num * 342.25, 2)

    elif 'paisa' in val:
        # 4 paisa = 1 aana = 342.25 sqft
        # Example: "3 paisa" → (3/4) × 342.25 = 256.69 sqft
        return round((num / 4) * 342.25, 2)

    elif 'ropani' in val:
        # 1 ropani = 16 aana = 16 × 342.25 sqft
        # Example: "1 ropani" → 5476 sqft
        return round(num * 16 * 342.25, 2)

    else:
        # No unit found — plain number like "1369"
        # Assume sqft as that's the most common unit
        return num

# Create new column — keep original intact
df_lph['built_up_sqft'] = df_lph['built_up_area'].apply(parse_built_up_area)

# Verify
print("=== built_up_sqft stats ===")
print(df_lph['built_up_sqft'].describe())
print(f"\nNulls: {df_lph['built_up_sqft'].isnull().sum()}")
print(f"Below 100 sqft:   {(df_lph['built_up_sqft'] < 100).sum()}")
print(f"Above 15000 sqft: {(df_lph['built_up_sqft'] > 15000).sum()}")

# Check suspicious values
print("\n=== Below 100 sqft ===")
print(df_lph[df_lph['built_up_sqft'] < 100][['built_up_area', 'built_up_sqft']].head(10))

print("\n=== Above 15000 sqft ===")
print(df_lph[df_lph['built_up_sqft'] > 15000][['built_up_area', 'built_up_sqft']].head(10))
```

```python
# ─────────────────────────────────────────
# STEP 6 — FIX year_built
# Two calendar systems mixed in data:
#   BS (Bikram Sambat) — Nepal calendar
#     Example: 2081, 2080, 2079 → clearly BS (>2025 impossible in AD)
#   AD (Anno Domini) — Western calendar
#     Example: 2024, 2023, 2022 → could be AD or BS
# Rule:
#   If year > 2025 → definitely BS → convert to AD by subtracting 56
#   If year <= 2025 → could be AD already → keep as is
#   If year < 1950 → too old, suspicious → NaN
# Example:
#   2081 BS → 2081 - 56 = 2025 AD ✅
#   2060 BS → 2060 - 56 = 2004 AD ✅
#   2024 AD → keep as 2024 AD ✅
# ─────────────────────────────────────────

print("=== Raw year_built distribution ===")
print(df_lph['year_built'].value_counts().sort_index().to_string())
```

```python
# ─────────────────────────────────────────
# STEP 6 — FIX year_built
# ─────────────────────────────────────────

def parse_year_built(val):
    if pd.isna(val):
        return np.nan

    val = str(val).strip()

    # Remove all spaces for easier matching
    val_clean = val.replace(' ', '').upper()

    # Garbage cases — return NaN
    # "2062+2079" has + sign, "20" too short, "79" too short
    if '+' in val_clean:
        return np.nan
    if len(re.sub(r'[^0-9]', '', val_clean)) < 4:
        return np.nan
    if val_clean in ['20222']:
        return np.nan

    # Extract the 4-digit year number
    num_match = re.search(r'(\d{4})', val_clean)
    if not num_match:
        return np.nan
    year = int(num_match.group(1))

    # Explicitly marked as AD — keep as is
    # Example: "1975 AD", "2007 AD", "2022 A.D", "2010 A.D"
    if 'AD' in val_clean or 'A.D' in val_clean:
        return year

    # Explicitly marked as BS — subtract 56
    # Example: "2050 BS", "2076 BS", "B.S 2076", "2021 AS" (typo for BS)
    if 'BS' in val_clean or 'B.S' in val_clean or 'AS' in val_clean:
        return year - 56

    # Now handle plain numbers by range
    if year > 2025:
        # Impossible to be AD (future) — must be BS
        # Example: 2081 BS → 2081 - 56 = 2025 AD
        return year - 56
    elif 1950 <= year <= 2025:
        # Could be AD — treat as AD directly
        # Example: 2024 → 2024 AD
        return year
    else:
        # Too old or weird — NaN
        return np.nan

df_lph['year_built_ad'] = df_lph['year_built'].apply(parse_year_built)
```

```python
# Verify
print("=== year_built_ad stats ===")
print(df_lph['year_built_ad'].describe())
print(f"\nNulls: {df_lph['year_built_ad'].isnull().sum()}")
print(f"Before 1990: {(df_lph['year_built_ad'] < 1990).sum()}")
print(f"After 2025:  {(df_lph['year_built_ad'] > 2025).sum()}")
print(f"\n=== Distribution ===")
print(df_lph['year_built_ad'].value_counts().sort_index().to_string())
```

```python
from datetime import datetime

# ─────────────────────────────────────────
# FIX YEAR_BUILT EDGE CASES
# Why: 2026-2033 are future years — these are
# under construction properties listed in advance
# Cap at current year since we can't verify completion
# Why keep pre-1990: Old houses are valid listings
# just rare in Kathmandu valley market
# ─────────────────────────────────────────

# Get current year dynamically
current_year = datetime.now().year
print(f"Current year: {current_year}")

# Cap future years at current year
# Why: Under construction listings show future completion year
# We cap at current year since house doesn't exist yet
df_lph['year_built_ad'] = df_lph['year_built_ad'].clip(upper=current_year)

# Calculate house age from current year
# Why age over raw year: Model learns "5 year old house costs X"
# which generalizes better than "house built in 2020 costs X"
# Example: year_built=2020, current_year=2026 → age=6
df_lph['house_age'] = current_year - df_lph['year_built_ad']

print("\n=== house_age stats ===")
print(df_lph['house_age'].describe())
print(f"\nNulls: {df_lph['house_age'].isnull().sum()}")
print(f"Age 0 (brand new):  {(df_lph['house_age'] == 0).sum()}")
print(f"Age above 40:       {(df_lph['house_age'] > 40).sum()}")
print(f"\nMost common ages:")
print(df_lph['house_age'].value_counts().sort_index().head(15))
```

```python
# ─────────────────────────────────────────
# STEP 7 — PARSE road_access → road_width_feet
# Units found:
#   Feet → keep as is (main unit)
#   Meter → convert to feet (1 meter = 3.281 feet)
#   Decimal feet → keep as is e.g. "14.5 Feet" → 14.5
# Keeping original road_access column intact
# ─────────────────────────────────────────

def parse_road_access(val):
    if pd.isna(val):
        return np.nan

    val = str(val).strip().lower()

    # Extract numeric value
    num_match = re.search(r'(\d+\.?\d*)', val)
    if not num_match:
        return np.nan
    num = float(num_match.group(1))

    if 'meter' in val:
        # Convert meters to feet
        # 1 meter = 3.281 feet
        # Example: "20 Meter" → 20 × 3.281 = 65.62 feet
        # Example: "7.5 Meter" → 7.5 × 3.281 = 24.61 feet
        return round(num * 3.281, 1)
    else:
        # Already in feet — return directly
        # Example: "13 Feet" → 13, "14.5 Feet" → 14.5
        return num

# Create new column — keep original intact
df_lph['road_width_feet'] = df_lph['road_access'].apply(parse_road_access)

print("=== road_width_feet stats ===")
print(df_lph['road_width_feet'].describe())
print(f"\nNulls: {df_lph['road_width_feet'].isnull().sum()}")
print(f"Below 3 feet:   {(df_lph['road_width_feet'] < 3).sum()}")
print(f"Above 80 feet:  {(df_lph['road_width_feet'] > 80).sum()}")
print(f"\nValue distribution:")
print(df_lph['road_width_feet'].value_counts().head(10))
```

```python
# ─────────────────────────────────────────
# STEP 8 — CONVERT DISTANCE COLUMNS TO METERS
# Same function as lalpurja land dataset
# Handles: "1km" → 1000, "500m" → 500, "1.5km" → 1500
# Using re.search to safely extract numbers
# Keeping original columns intact
# ─────────────────────────────────────────

def convert_to_meters(val):
    if pd.isna(val):
        return np.nan

    val = str(val).strip().lower()

    # Use regex to extract first valid number
    # Safely handles formats like "1km", "500m", "1.5km", "2.5km"
    num_match = re.search(r'(\d+\.?\d*)', val)
    if not num_match:
        return np.nan
    num = float(num_match.group(1))

    if 'km' in val:
        # Multiply by 1000 to convert km → meters
        # Example: "1km" → 1000, "1.5km" → 1500
        return num * 1000
    elif 'm' in val:
        # Already in meters — return as is
        # Example: "500m" → 500, "200m" → 200
        return num

    return np.nan

# Distance columns in this dataset
# Note: ring_road is also a distance column
distance_cols = ['hospital', 'airport', 'pharmacy', 'bhatbhateni',
                 'school', 'college', 'gym', 'public_transport',
                 'police_station', 'pashupatinath', 'boudhanath',
                 'atm', 'hotel', 'restaurant', 'banquet', 'ring_road']

for col in distance_cols:
    df_lph[col + '_m'] = df_lph[col].apply(convert_to_meters)
    print(f"✅ {col} → {col}_m converted")

print("\n=== Sample converted distances ===")
print(df_lph[['hospital_m', 'airport_m', 'school_m', 'ring_road_m']].describe().round(0))

print(f"\n=== Nulls after conversion ===")
new_dist_cols = [col + '_m' for col in distance_cols]
nulls = df_lph[new_dist_cols].isnull().sum()
print(nulls)
```

```python
# ─────────────────────────────────────────
# CAP UNREALISTIC DISTANCE VALUES
# Same approach as lalpurja land dataset
# Realistic maximum distances for Kathmandu valley
# ─────────────────────────────────────────
distance_caps = {
    'hospital_m':          15000,   # 15km max
    'airport_m':           30000,   # 30km max
    'pharmacy_m':          10000,   # 10km max
    'bhatbhateni_m':       15000,   # 15km max
    'school_m':            10000,   # 10km max
    'college_m':           15000,   # 15km max
    'gym_m':               10000,   # 10km max
    'public_transport_m':   5000,   # 5km max
    'police_station_m':    15000,   # 15km max
    'pashupatinath_m':     30000,   # 30km max
    'boudhanath_m':        30000,   # 30km max
    'atm_m':                5000,   # 5km max
    'hotel_m':             15000,   # 15km max
    'restaurant_m':        10000,   # 10km max
    'banquet_m':           15000,   # 15km max
    'ring_road_m':         30000,   # 30km max
}

for col, cap in distance_caps.items():
    over_cap = (df_lph[col] > cap).sum()
    if over_cap > 0:
        print(f"⚠️  {col}: {over_cap} rows exceed {cap}m — capping")
    df_lph.loc[df_lph[col] > cap, col] = np.nan

print("\n=== Distance stats after capping ===")
key_cols = ['hospital_m', 'airport_m', 'school_m',
            'ring_road_m', 'public_transport_m']
print(df_lph[key_cols].describe().round(0))

# ─────────────────────────────────────────
# DECISION ON HIGH-NULL DISTANCE COLUMNS
# gym, atm, hotel, restaurant, banquet
# have 800-900 nulls out of 2190 rows (40%+)
# These are less critical amenities for house pricing
# Check if they add value before deciding to keep or drop
# ─────────────────────────────────────────
new_dist_cols = [col + '_m' for col in distance_cols]
print("\n=== Null % per distance column ===")
for col in new_dist_cols:
    null_pct = df_lph[col].isnull().sum() / len(df_lph) * 100
    print(f"{col}: {null_pct:.1f}%")
```

```python
# ─────────────────────────────────────────
# CATEGORY 7 — ROW DELETION
# Why: total_price and property_type are critical
# We cannot guess the target variable (total_price)
# and property_type affects everything else
# Dropping these rows is safer than imputing them
# ─────────────────────────────────────────
before = len(df_lph)

df_lph.dropna(subset=['total_price', 'property_type'], inplace=True)
df_lph.reset_index(drop=True, inplace=True)

print(f"Rows before: {before}")
print(f"Rows after:  {len(df_lph)}")
print(f"Rows dropped: {before - len(df_lph)}")
print(f"\n=== Remaining nulls ===")
nulls = df_lph.isnull().sum()
print(nulls[nulls > 0].sort_values(ascending=False))
```

```python
# ─────────────────────────────────────────
# CATEGORY 1 — DROP HIGH NULL COLUMNS
# Why: Columns with 40%+ nulls contain more
# fabricated data than real data after imputation
# This makes the model learn false patterns
# Example: gym_m is 38% null — if we fill with median
# we're essentially making up values for 800 rows
# The model might think gym distance matters when
# it's actually just our made-up median repeated
# ─────────────────────────────────────────

drop_cols = [
    # High null amenities — too many missing to impute reliably
    'dimension',     # 63.7% null — already parsed into facing_road_width in land
                     # but housing dimension means something different — drop
    'banquet',       # 40.0% null — not a strong house price predictor
    'hotel',         # 39.4% null — not a strong house price predictor
    'atm',           # 38.8% null — not a strong house price predictor
    'restaurant',    # 38.6% null — not a strong house price predictor
    'gym',           # 36.5% null — not a strong house price predictor

    # Original string distance columns
    # Already converted to _m versions — raw strings no longer needed
    'police_station', 'ring_road', 'college', 'bhatbhateni',
    'pharmacy', 'boudhanath', 'pashupatinath', 'public_transport',
    'hospital', 'airport', 'school',

    # Other raw columns already parsed into cleaner versions
    'road_access',    # → road_width_feet
    'built_up_area',  # → built_up_sqft
    'year_built',     # → house_age
    'year_built_ad',  # → house_age
]

df_lph.drop(columns=drop_cols, inplace=True)

print(f"Shape after dropping: {df_lph.shape}")
print(f"\n=== Remaining columns ===")
print(list(df_lph.columns))

print(f"\n=== Remaining nulls ===")
nulls = df_lph.isnull().sum()
print(nulls[nulls > 0].sort_values(ascending=False))
```

```python
# ─────────────────────────────────────────
# CATEGORY 3 — CONDITIONAL FILLING
# Why conditional: Some nulls have a logical
# real-world default rather than a statistical one
# ─────────────────────────────────────────

# furnishing → fill with 'Unfurnished'
# Why: In Nepal real estate, if furnishing is not
# mentioned it typically means unfurnished
# It's the conservative safe default
# Example: Listing says nothing about furniture
#          → assume buyer gets empty house → Unfurnished
df_lph['furnishing'] = df_lph['furnishing'].fillna('Unfurnished')
print(f"✅ furnishing filled — distribution:")
print(df_lph['furnishing'].value_counts())

# negotiable → fill with 'Yes'
# Why: We confirmed earlier that 90%+ listings
# are negotiable in Nepal housing market
# NaN likely means seller forgot to mention it
# not that it's non-negotiable
# Example: 1310 out of 1326 lalpurja land were Yes
#          → safe to assume missing = Yes
df_lph['negotiable'] = df_lph['negotiable'].fillna('Yes')
print(f"\n✅ negotiable filled — distribution:")
print(df_lph['negotiable'].value_counts())

# road_type → fill with 'High Access'
# Why: 12 nulls only (0.5%) — very small
# High Access is the most common road type (75%)
# Most Kathmandu urban listings have blacktopped roads
df_lph['road_type'] = df_lph['road_type'].fillna('High Access')
print(f"\n✅ road_type filled — distribution:")
print(df_lph['road_type'].value_counts())

print(f"\n=== Nulls after Category 3 ===")
nulls = df_lph.isnull().sum()
print(nulls[nulls > 0].sort_values(ascending=False))
```

```python
# ─────────────────────────────────────────
# CATEGORY 2 — LOGIC BASED DEDUCTION
# How it works: Geography is hierarchical in Nepal
#   neighborhood → municipality → district
# If we know a property is in "Budhanilkantha"
# we can reliably infer its municipality is
# "Budhanilkantha" and district is "Kathmandu"
# This is more accurate than statistical imputation
# because it's based on actual geographic facts
#
# Example:
#   Row has neighborhood="Imadol", municipality=NaN
#   Other rows with neighborhood="Imadol" have
#   municipality="Mahalaxmi" (most common)
#   → fill NaN with "Mahalaxmi" ✅ geographically correct
# ─────────────────────────────────────────

# FIX — Convert ward_no to numeric first

df_lph['ward_no'] = pd.to_numeric(df_lph['ward_no'], errors='coerce')
print(f"ward_no dtype: {df_lph['ward_no'].dtype}")
print(f"ward_no nulls after conversion: {df_lph['ward_no'].isnull().sum()}")

# Step 1 — Fill neighborhood first (only 2 nulls)
# Use district mode as fallback — very few nulls
df_lph['neighborhood'] = df_lph.groupby('district')['neighborhood']\
    .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
print(f"✅ neighborhood nulls filled: {df_lph['neighborhood'].isnull().sum()}")

# Step 2 — Fill municipality using neighborhood mode
# Why neighborhood: Properties in same neighborhood
# almost always belong to same municipality
# Example: All "Imadol" properties → "Mahalaxmi"
df_lph['municipality'] = df_lph.groupby('neighborhood')['municipality']\
    .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan))

# Fill remaining using district mode as fallback
df_lph['municipality'] = df_lph.groupby('district')['municipality']\
    .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
print(f"✅ municipality nulls filled: {df_lph['municipality'].isnull().sum()}")

# Step 3 — Fill ward_no using neighborhood mode
# Why neighborhood: Ward numbers are geographically
# consistent — all properties in "Imadol" are
# typically in the same ward (Ward 3 of Mahalaxmi)
# Example:
#   Row has neighborhood="Imadol", ward_no=NaN
#   Other "Imadol" rows have ward_no=3 (most common)
#   → fill with 3 ✅ geographically accurate
df_lph['ward_no'] = df_lph.groupby('neighborhood')['ward_no']\
    .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan))

# Fill remaining using municipality median as fallback
df_lph['ward_no'] = df_lph.groupby('municipality')['ward_no']\
    .transform(lambda x: x.fillna(x.median() if not x.empty else np.nan))

# Fill any last remaining with global median
df_lph['ward_no'] = df_lph['ward_no'].fillna(df_lph['ward_no'].median())
print(f"✅ ward_no nulls filled: {df_lph['ward_no'].isnull().sum()}")

print(f"\n=== Nulls after Category 2 ===")
nulls = df_lph.isnull().sum()
print(nulls[nulls > 0].sort_values(ascending=False))
```

```python
from sklearn.impute import KNNImputer
```

```python
# ─────────────────────────────────────────
# CATEGORY 4 — GROUPED MEDIAN IMPUTATION
# How it works: Instead of filling with global median
# we group by neighborhood first and use that group's median
# Why: A missing road width in Baluwatar is likely 16-20ft
# but in a rural area it might be 10ft
# Neighborhood-level median captures this local difference
# Example:
#   road_width_feet missing for a property in "Imadol"
#   All other "Imadol" properties have road_width median=13ft
#   → fill with 13ft ✅ much more accurate than global 15.5ft
# ─────────────────────────────────────────

# property_face — neighborhood mode then global mode
# Why mode not median: property_face is categorical not numeric
# Most common facing direction in a neighborhood is best guess
df_lph['property_face'] = df_lph.groupby('neighborhood')['property_face']\
    .transform(lambda x: x.fillna(x.mode()[0] if len(x.mode()) > 0 else np.nan))
df_lph['property_face'] = df_lph['property_face']\
    .fillna(df_lph['property_face'].mode()[0])
print(f"✅ property_face filled: {df_lph['property_face'].isnull().sum()} nulls remaining")

# road_width_feet — neighborhood median then global median
df_lph['road_width_feet'] = df_lph.groupby('neighborhood')['road_width_feet']\
    .transform(lambda x: x.fillna(x.median()))
df_lph['road_width_feet'] = df_lph['road_width_feet']\
    .fillna(df_lph['road_width_feet'].median())
print(f"✅ road_width_feet filled: {df_lph['road_width_feet'].isnull().sum()} nulls remaining")

# land_size_aana — neighborhood median then global median
df_lph['land_size_aana'] = df_lph.groupby('neighborhood')['land_size_aana']\
    .transform(lambda x: x.fillna(x.median()))
df_lph['land_size_aana'] = df_lph['land_size_aana']\
    .fillna(df_lph['land_size_aana'].median())
print(f"✅ land_size_aana filled: {df_lph['land_size_aana'].isnull().sum()} nulls remaining")
```

```python
# CATEGORY 5 — GROUPED MEDIAN IMPUTATION


# built_up_sqft — neighborhood + land_size bin grouped median
# Why both: Built up area depends on plot size AND location
# A 4 aana plot in Baluwatar has different built up than 4 aana in Bhaktapur
# Example:
#   Missing built_up_sqft for 4 aana plot in "Imadol"
#   Other "Imadol" + "small plot" properties have median=1026 sqft
#   → fill with 1026 sqft ✅ more accurate than global 1197 sqft
df_lph['land_size_bin'] = pd.cut(df_lph['land_size_aana'],
                                  bins=[0, 3, 5, 8, 15, 100],
                                  labels=['tiny', 'small', 'medium', 'large', 'xlarge'])
df_lph['built_up_sqft'] = df_lph.groupby(['neighborhood', 'land_size_bin'])['built_up_sqft']\
    .transform(lambda x: x.fillna(x.median()))
df_lph['built_up_sqft'] = df_lph.groupby('neighborhood')['built_up_sqft']\
    .transform(lambda x: x.fillna(x.median()))
df_lph['built_up_sqft'] = df_lph['built_up_sqft']\
    .fillna(df_lph['built_up_sqft'].median())
df_lph.drop(columns=['land_size_bin'], inplace=True)
print(f"✅ built_up_sqft filled: {df_lph['built_up_sqft'].isnull().sum()} nulls remaining")
```

```python
# ─────────────────────────────────────────
# CATEGORY 6 — TEMPORAL IMPUTATION
# How it works: Newer neighborhoods have newer houses
# Using neighborhood median year is more accurate
# than global median because construction boomed
# in different areas at different times
# Example:
#   Anamnaagr (Kathmandu) — mostly older houses, median age=12
#   Imadol (Lalitpur) — newer development, median age=3
#   → fill missing house_age using neighborhood median
#   → much better than global median of 7.8 years
# ─────────────────────────────────────────
df_lph['house_age'] = df_lph.groupby('neighborhood')['house_age']\
    .transform(lambda x: x.fillna(x.median()))
df_lph['house_age'] = df_lph['house_age']\
    .fillna(df_lph['house_age'].median())
print(f"✅ house_age filled: {df_lph['house_age'].isnull().sum()} nulls remaining")
```

```python
# ─────────────────────────────────────────
# CATEGORY 5 — KNN IMPUTATION FOR DISTANCES
# How it works:
# KNN = K-Nearest Neighbors
# Step 1: Look at all distance columns together
# Step 2: For a row with missing hospital_m,
#         find K=5 most similar rows based on
#         OTHER distance values that ARE present
# Step 3: Use average of those 5 rows' hospital_m
#         as the imputed value
#
# Why KNN for distances:
# Properties near each other share similar distances
# to landmarks — if two properties have similar
# airport_m and ring_road_m, they probably also have
# similar hospital_m even if one is missing
#
# Example:
#   Property A: airport_m=5000, ring_road_m=2000, hospital_m=NaN
#   5 most similar properties (by airport+ring_road) have:
#   hospital_m = [800, 900, 750, 1000, 850]
#   → imputed hospital_m = mean(800,900,750,1000,850) = 860m ✅
#   Much better than global median of 1000m
#
# n_neighbors=5: use 5 most similar properties
# More neighbors = smoother but less local
# Fewer neighbors = more local but noisier
# 5 is a good balance for this dataset size
# ─────────────────────────────────────────
dist_knn_cols = ['hospital_m', 'airport_m', 'pharmacy_m', 'bhatbhateni_m',
                 'school_m', 'college_m', 'public_transport_m', 'police_station_m',
                 'pashupatinath_m', 'boudhanath_m', 'ring_road_m']

knn_imputer = KNNImputer(n_neighbors=5)

# KNN works on numeric data only
# It finds similarity based on non-null values in the row
# then uses those similar rows to estimate missing values
df_lph[dist_knn_cols] = knn_imputer.fit_transform(df_lph[dist_knn_cols])
print(f"✅ Distance columns KNN imputed")
for col in dist_knn_cols:
    print(f"   {col}: {df_lph[col].isnull().sum()} nulls remaining")
```

```python
# ─────────────────────────────────────────
# FIX REMAINING ISSUES
# ─────────────────────────────────────────

# Fix 1 — property_face lowercase duplicates
# Why: "west" and "West" are same direction
# but model treats them as different categories
# Title case standardizes all values
df_lph['property_face'] = df_lph['property_face'].str.title()
print("=== property_face after fixing ===")
print(df_lph['property_face'].value_counts())

# Fix 2 — land_size_aana
# Cap minimum at 1 aana — smallest realistic residential plot
# Cap maximum at 50 aana — consistent with lalpurja land model
# Why 1 minimum: 0 aana means no land — impossible for a house
df_lph['land_size_aana'] = df_lph['land_size_aana'].clip(lower=1, upper=50)
print(f"\n✅ land_size_aana clipped: min={df_lph['land_size_aana'].min()}, max={df_lph['land_size_aana'].max()}")

# Fix 3 — built_up_sqft
# Replace values below 100 sqft with neighborhood median
# Why: 2.5 sqft is impossible for a house — clearly wrong
# Using neighborhood median preserves local building patterns
mask = df_lph['built_up_sqft'] < 100
df_lph.loc[mask, 'built_up_sqft'] = df_lph.loc[mask, 'neighborhood']\
    .map(df_lph.groupby('neighborhood')['built_up_sqft']\
    .apply(lambda x: x[x >= 100].median()))
# Fill any remaining with global median
df_lph['built_up_sqft'] = df_lph['built_up_sqft']\
    .fillna(df_lph['built_up_sqft'].median())
print(f"✅ built_up_sqft fixed: min={df_lph['built_up_sqft'].min():.1f}, max={df_lph['built_up_sqft'].max():.1f}")

# Fix 4 — house_age cap at 60
# Why: 61 year old house is suspicious in Kathmandu
# Most older houses have been rebuilt
# Cap at 60 to remove extreme outlier
df_lph['house_age'] = df_lph['house_age'].clip(upper=60)
print(f"✅ house_age capped: max={df_lph['house_age'].max()}")

# Fix 5 — road_width_feet cap at 80 feet
# Why: 100 feet road is extreme for residential areas
# Cap at 80 feet — realistic maximum for urban Nepal
df_lph['road_width_feet'] = df_lph['road_width_feet'].clip(upper=80)
print(f"✅ road_width_feet capped: max={df_lph['road_width_feet'].max()}")

# ─────────────────────────────────────────
# FINAL VERIFY
# ─────────────────────────────────────────
print(f"\n=== Final shape: {df_lph.shape} ===")
print(f"Nulls: {df_lph.isnull().sum().sum()}")
print(f"Duplicates: {df_lph.duplicated().sum()}")
print(f"price range: {df_lph['total_price'].min():,.0f} – {df_lph['total_price'].max():,.0f}")
print(f"land range: {df_lph['land_size_aana'].min()} – {df_lph['land_size_aana'].max()} aana")
print(f"built_up range: {df_lph['built_up_sqft'].min():.0f} – {df_lph['built_up_sqft'].max():.0f} sqft")
```

```python
# ─────────────────────────────────────────
# COLUMN DESCRIPTIONS — LALPURJA HOUSING DATASET
# ─────────────────────────────────────────

# ── LOCATION COLUMNS ───────────────────────────────────────────────────────

# district
# Which of the 3 Kathmandu valley districts the property is in
# Values: Kathmandu (1490), Lalitpur (525), Bhaktapur (175)
# Example: "Kathmandu"

# municipality
# Local government body the property falls under
# More granular than district — captures micro-market differences
# Example: "Budhanilkantha", "Mahalaxmi", "Suryabinayak"
# Filled using neighborhood-level mode (Category 2 imputation)

# ward_no
# Ward number within the municipality (1–33)
# Most granular location identifier available
# Example: Ward 3 of Mahalaxmi municipality = Imadol area
# Filled using neighborhood-level mode (Category 2 imputation)

# neighborhood
# Neighborhood name extracted from listing
# Key price signal — location within ward level
# Example: "Imadol", "Budhanilkantha", "Balkot"
# Only 2 nulls — filled with district mode

# ── PROPERTY CHARACTERISTICS ───────────────────────────────────────────────

# property_type
# Land use classification of the property
# Values: Residential (1876), Commercial (190), Semi-commercial (124)
# Example: "Residential"

# property_face
# Cardinal direction the property faces
# Values: East (528), South (332), West (314), South-East (283) etc.
# Standardized to title case — "west" → "West"
# Vastu shastra affects buyer preference in Nepal
# Example: "East" facing is most preferred

# furnishing
# Furniture status of the property
# Values: Semi Furnished (1077), Unfurnished (792), Full Furnished (321)
# NaN filled with "Unfurnished" — conservative default for Nepal market
# Example: "Semi Furnished"

# road_type
# Quality of road accessing the property
# Values: High Access (1655) = Blacktopped/Paved
#         Low Access (535)  = Gravelled/Soil/Alley
# NaN filled with "High Access" — most common in urban Nepal
# Example: "High Access"

# negotiable
# Whether the listed price is open to negotiation
# Values: Yes (2155), No (35)
# 98.4% are Yes — NaN filled with "Yes" (Nepal market standard)
# Example: "Yes"

# ── SIZE COLUMNS ───────────────────────────────────────────────────────────

# land_size_aana
# Size of the plot/land in aana (1 aana ≈ 31.8 sqm)
# Parsed from property_area — handled aana/ropani/sqft/paisa/daam units
# Clipped: min=1 aana, max=50 aana
# Example: "4 Aana" → 4.0

# built_up_sqft
# Built up area of the house in square feet
# Parsed from built_up_area — handled sqft/aana/sqm units
# Mislabeled aana values (>50) treated as sqft
# Example: "1369 square-feet" → 1369.0
#          "3 aana" → 3 × 342.25 = 1026.75 sqft

# house_age
# Age of the house in years (current_year - year_built)
# Calculated dynamically using datetime.now().year
# BS calendar years converted to AD (subtract 56)
# Example: year_built=2079 BS → 2023 AD → age=3 years
# Filled with neighborhood median (newer areas = newer houses)

# road_width_feet
# Width of the road accessing the property in feet
# Parsed from road_access — handled feet and meter units
# "20 Meter" → 20 × 3.281 = 65.62 feet
# Capped at 80 feet maximum
# Example: "13 Feet" → 13.0

# ── LISTING METADATA ───────────────────────────────────────────────────────

# property_area
# Original raw land size string — kept for reference
# Already parsed into land_size_aana
# Example: "4 Aana", "1 Ropani", "2000 Sq. Ft."
# Will be dropped before modeling

# date_posted
# Date the listing was posted on lalpurja.com
# Format: "2025 November 8"
# Not used for modeling — kept for reference
# Will be dropped before modeling

# views
# Number of times the listing was viewed on lalpurja.com
# Near-zero correlation with price (confirmed in lalpurja land EDA)
# Will likely be dropped before modeling

# ── TARGET VARIABLE ────────────────────────────────────────────────────────

# total_price
# Total asking price of the property in NPR
# Parsed from total_price_raw — "3 Crore 50 Lakh" → 35,000,000 NPR
# Range: 10,000,000 (1 Crore) to 350,000,000 (35 Crore)
# This is the TARGET variable for regression modeling

# ── AMENITY DISTANCE COLUMNS (all in meters) ───────────────────────────────

# hospital_m
# Distance to nearest hospital in meters
# Capped at 15,000m — KNN imputed for nulls
# Example: "1km" → 1000m, "500m" → 500m

# airport_m
# Distance to Tribhuvan International Airport in meters
# Strong price predictor — closer to airport = more urban = higher price
# Capped at 30,000m — KNN imputed for nulls
# Example: "6km" → 6000m

# pharmacy_m
# Distance to nearest pharmacy in meters
# Proxy for urban infrastructure density
# Capped at 10,000m — KNN imputed for nulls

# bhatbhateni_m
# Distance to nearest Bhatbhateni Supermarket in meters
# Major landmark supermarket chain — proxy for urban amenity access
# Capped at 15,000m — KNN imputed for nulls

# school_m
# Distance to nearest school in meters
# Important for families — affects residential demand
# Capped at 10,000m — KNN imputed for nulls

# college_m
# Distance to nearest college in meters
# More relevant for semi-commercial and rental properties
# Capped at 15,000m — KNN imputed for nulls
# Had 29% nulls — higher than other amenities

# public_transport_m
# Distance to nearest public transport stop in meters
# Bus stop, microbus stand — measures daily commute convenience
# Capped at 5,000m — KNN imputed for nulls

# police_station_m
# Distance to nearest police station in meters
# Proxy for safety and urban infrastructure
# Had 31.7% nulls — highest among kept amenity columns
# Capped at 15,000m — KNN imputed for nulls

# pashupatinath_m
# Distance to Pashupatinath Temple in meters
# Major geographic landmark — captures centrality within valley
# Capped at 30,000m — KNN imputed for nulls

# boudhanath_m
# Distance to Boudhanath Stupa in meters
# Major geographic landmark — reference point for east Kathmandu
# Capped at 30,000m — KNN imputed for nulls

# ring_road_m
# Distance to Kathmandu Ring Road in meters
# Strong price predictor — ring road proximity = urban = higher price
# Had 30.7% nulls — KNN imputed using spatial similarity
# Capped at 30,000m
# Example: "2km" → 2000m
```
