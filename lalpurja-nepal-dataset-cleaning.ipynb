```python
import numpy as np    
import pandas as pd
```

```python
# ─────────────────────────────────────────
# LOAD
# ─────────────────────────────────────────
ktm = pd.read_csv('lalpurja_kathmandu_land_fixed.csv')
lat = pd.read_csv('lalpurja_lalitpur_land_fixed.csv')
bkt = pd.read_csv('lalpurja_bhaktapur_land_fixed.csv')
```

```python
# ─────────────────────────────────────────
# ANALYZE EACH FILE INDIVIDUALLY
# ─────────────────────────────────────────
for name, df in [('Kathmandu', ktm), ('Lalitpur', lat), ('Bhaktapur', bkt)]:
    print(f"\n{'='*50}")
    print(f"=== {name} ===")
    print(f"Shape: {df.shape}")

    print(f"\n--- pricing_raw ---")
    print(df['pricing_raw'].value_counts().head(8))

    print(f"\n--- property_area ---")
    print(df['property_area'].value_counts().head(8))

    print(f"\n--- dimension ---")
    print(df['dimension'].value_counts().head(8))

    print(f"\n--- road_access ---")
    print(df['road_access'].value_counts().head(8))

    print(f"\n--- property_type ---")
    print(df['property_type'].value_counts())

    print(f"\n--- negotiable ---")
    print(df['negotiable'].value_counts())

    print(f"\n--- hospital sample ---")
    print(df['hospital'].value_counts().head(5))

    print(f"\n--- Nulls ---")
    nulls = df.isnull().sum()
    print(nulls[nulls > 0])
```

```python
# Assign correct district names
ktm['district'] = 'Kathmandu'
lat['district'] = 'Lalitpur'
bkt['district'] = 'Bhaktapur'

# Combine
df_lp = pd.concat([ktm, lat, bkt], ignore_index=True)

print(f"Combined shape: {df_lp.shape}")
print(f"Districts: {df_lp['district'].value_counts().to_dict()}")
print(f"Total nulls: {df_lp.isnull().sum().sum()}")
```

```python
# ─────────────────────────────────────────
# STEP 1 — DROP USELESS COLUMNS
# ─────────────────────────────────────────
drop_cols = ['url', 'property_id', 'source', 'purpose',
             'city_area', 'ward_office', 'title']
df_lp.drop(columns=drop_cols, inplace=True)
print(f"After dropping: {df_lp.shape}")
```

```python
# ─────────────────────────────────────────
# STEP 2 — DROP NULL ROWS
# ─────────────────────────────────────────
df_lp.dropna(subset=['pricing_raw', 'property_area'], inplace=True)
df_lp.reset_index(drop=True, inplace=True)
print(f"After dropping null rows: {df_lp.shape}")
```

```python
# ─────────────────────────────────────────
# STEP 3 — PARSE pricing_raw
# Extract numeric lakh value and filter Per Aana only
# ─────────────────────────────────────────
# Keep only Per Aana listings
df_lp = df_lp[df_lp['pricing_raw'].str.contains('Per Aana', na=False)].copy()
print(f"After keeping Per Aana only: {df_lp.shape}")

# Extract numeric value and convert to NPR
df_lp['price_per_aana'] = df_lp['pricing_raw'].str.extract(r'(\d+\.?\d*)').astype(float)
df_lp['price_per_aana'] = df_lp['price_per_aana'] * 100_000  # lakh to NPR

print("\n=== price_per_aana sample ===")
print(df_lp['price_per_aana'].describe())
```

```python
# ─────────────────────────────────────────
# STEP 4 — PARSE property_area
# ─────────────────────────────────────────
df_lp['land_size_aana'] = df_lp['property_area'].str.extract(r'(\d+\.?\d*)').astype(float)

print("\n=== land_size_aana sample ===")
print(df_lp['land_size_aana'].describe())
```

```python
# ─────────────────────────────────────────
# STEP 5 — PARSE road_access (actual road width)
# ─────────────────────────────────────────
df_lp['road_width_feet'] = df_lp['road_access'].str.extract(r'(\d+\.?\d*)').astype(float)

print("\n=== road_width_feet sample ===")
print(df_lp['road_width_feet'].describe())
```

```python
# ─────────────────────────────────────────
# STEP 6 — PARSE dimension (facing road width)
# Handle formats: F 28 Feet, F55feet, F 78 | F 80 Feet
# ─────────────────────────────────────────
df_lp['facing_road_width'] = df_lp['dimension'].str.extract(r'F\s*(\d+)').astype(float)

print("\n=== facing_road_width sample ===")
print(df_lp['facing_road_width'].describe())
```

```python
# ─────────────────────────────────────────
# STEP 7 — VERIFY
# ─────────────────────────────────────────
print("\n=== Nulls after parsing ===")
nulls = df_lp.isnull().sum()
print(nulls[nulls > 0])
print(f"\nShape: {df_lp.shape}")
```

```python
import re

def convert_to_meters(val):
    # If value is NaN/None, return NaN directly
    if pd.isna(val):
        return np.nan
    
    # Convert to lowercase string and remove whitespace
    val = str(val).strip().lower()
    
    # Use regex to extract the first valid number from the string
    # This handles cases like "1km", "500m", "1.5km", "2.5km" safely
    # re.search finds the first occurrence of digits with optional decimal
    num_match = re.search(r'(\d+\.?\d*)', val)
    
    # If no number found at all, return NaN
    if not num_match:
        return np.nan
    
    # Extract the matched number and convert to float
    num = float(num_match.group(1))
    
    if 'km' in val:
        # Multiply by 1000 to convert kilometers to meters
        return num * 1000
    elif 'm' in val:
        # Already in meters, return as is
        return num
    
    # If neither km nor m found, return NaN
    return np.nan

# Apply to all distance columns
distance_cols = ['hospital', 'airport', 'pharmacy', 'bhatbhateni',
                 'school', 'public_transport', 'police_station',
                 'pashupatinath', 'boudhanath', 'ring_road']

for col in distance_cols:
    df_lp[col + '_m'] = df_lp[col].apply(convert_to_meters)
    print(f"✅ {col} → {col}_m converted")

# Drop original string distance columns
df_lp.drop(columns=distance_cols, inplace=True)

print("\n=== Sample converted distances (meters) ===")
print(df_lp[['hospital_m', 'airport_m', 'school_m', 'ring_road_m']].describe())

print(f"\n=== Nulls after conversion ===")
nulls = df_lp.isnull().sum()
print(nulls[nulls > 0])

print(f"\nShape: {df_lp.shape}")
```

```python
# ─────────────────────────────────────────
# CHECK SUSPICIOUS EXTREME VALUES
# ─────────────────────────────────────────
# For urban Kathmandu valley, realistic distances are:
# hospital: 100m – 10km (10,000m)
# airport: 2km – 25km (25,000m)
# school: 50m – 5km (5,000m)
# ring_road: 100m – 20km (20,000m)

print("=== Rows with hospital_m > 10000 ===")
print(df_lp[df_lp['hospital_m'] > 10000][['hospital_m', 'neighborhood']].head(10))

print("\n=== Rows with school_m > 5000 ===")
print(df_lp[df_lp['school_m'] > 5000][['school_m', 'neighborhood']].head(10))

print("\n=== Rows with ring_road_m > 20000 ===")
print(df_lp[df_lp['ring_road_m'] > 20000][['ring_road_m', 'neighborhood']].head(10))
```

```python
# ─────────────────────────────────────────
# CAP UNREALISTIC DISTANCE VALUES
# ─────────────────────────────────────────
# Why: 3 rows have parsing errors producing values like 800,000m (800km)
# which are impossible for Kathmandu valley properties
# We cap at realistic maximum distances for each amenity
# rather than dropping rows — preserving valuable data

# Realistic max distances for Kathmandu valley (in meters)
distance_caps = {
    'hospital_m':        15000,   # 15km max
    'airport_m':         30000,   # 30km max
    'pharmacy_m':        10000,   # 10km max
    'bhatbhateni_m':     15000,   # 15km max
    'school_m':          10000,   # 10km max
    'public_transport_m': 5000,   # 5km max
    'police_station_m':  15000,   # 15km max
    'pashupatinath_m':   30000,   # 30km max
    'boudhanath_m':      30000,   # 30km max
    'ring_road_m':       30000,   # 30km max
}

for col, cap in distance_caps.items():
    # Count how many rows exceed the cap before fixing
    over_cap = (df_lp[col] > cap).sum()
    if over_cap > 0:
        print(f"⚠️ {col}: {over_cap} rows exceed {cap}m — capping")
    # Replace values above cap with NaN — they are parsing errors
    df_lp.loc[df_lp[col] > cap, col] = np.nan

# ─────────────────────────────────────────
# VERIFY
# ─────────────────────────────────────────
print("\n=== Distance stats after capping ===")
dist_cols = list(distance_caps.keys())
print(df_lp[dist_cols].describe().round(0))
```

```python
# ─────────────────────────────────────────
# STEP 9 — HANDLE REMAINING NULLS
# ─────────────────────────────────────────
# Check current null situation
print("=== Current nulls ===")
nulls = df_lp.isnull().sum()
print(nulls[nulls > 0])
```

```python
# ─────────────────────────────────────────
# FILL REMAINING NULLS
# ─────────────────────────────────────────

# Distance columns — fill with median
# Why median: distance data is skewed, median is more representative
# than mean which gets pulled by far away properties
dist_cols = [
    'hospital_m', 'airport_m', 'pharmacy_m', 'bhatbhateni_m',
    'school_m', 'public_transport_m', 'police_station_m',
    'pashupatinath_m', 'boudhanath_m', 'ring_road_m'
]
for col in dist_cols:
    median_val = df_lp[col].median()
    df_lp[col] = df_lp[col].fillna(median_val)
    
# Road width columns — fill with median
# Why median: road widths cluster at standard values (13ft, 20ft)
# median will naturally land on one of these standard values
df_lp['road_width_feet']   = df_lp['road_width_feet'].fillna(df_lp['road_width_feet'].median())
df_lp['facing_road_width'] = df_lp['facing_road_width'].fillna(df_lp['facing_road_width'].median())

# Categorical columns — fill with mode
# Why mode: most common value is the best guess for
# missing categorical entries like municipality and ward
for col in ['municipality', 'ward_no']:
    df_lp[col] = df_lp[col].fillna(df_lp[col].mode()[0])

# dimension column — drop remaining 4 nulls
# Why drop: dimension is a string column we already parsed into
# facing_road_width — no need to keep the original string
df_lp.drop(columns=['dimension'], inplace=True)

# ─────────────────────────────────────────
# VERIFY
# ─────────────────────────────────────────
print("=== Nulls after fixing ===")
nulls = df_lp.isnull().sum()
print(nulls[nulls > 0] if nulls[nulls > 0].any() else "✅ No nulls")

print(f"\nFinal shape: {df_lp.shape}")
print(f"\nColumns: {list(df_lp.columns)}")
```

```python
# ─────────────────────────────────────────
# APPLY FINAL FIXES
# ─────────────────────────────────────────

# Fix 1 — Remove suspicious price rows
# Why: Below 500k likely means parsing picked wrong number
# Above 50M is an extreme outlier for residential land
before = len(df_lp)
df_lp = df_lp[
    (df_lp['price_per_aana'] >= 500000) &
    (df_lp['price_per_aana'] <= 50000000)
].reset_index(drop=True)
print(f"After price filter: {len(df_lp)} rows (removed {before - len(df_lp)})")

# Fix 2 — Cap land size at 50 aana
# Why: Properties above 50 aana are likely commercial/agricultural
# not standard residential plots — same cap used in df_land
df_lp['land_size_aana'] = df_lp['land_size_aana'].clip(upper=50)
print(f"land_size_aana capped at 50 aana")

# Fix 3 — Standardize road_type
# Why: Need consistent categories to match df_land road types
# and for proper encoding in the ML model
road_map = {
    'Blacktopped':     'High Access',
    'Paved':           'High Access',
    'Gravelled':       'Low Access',
    'Soil Stabilized': 'Low Access',
    'alley':           'Low Access'
}
df_lp['road_type'] = df_lp['road_type'].replace(road_map)
print("\n=== road_type after standardization ===")
print(df_lp['road_type'].value_counts())

# Fix 4 — Drop columns not needed for modeling
# Why: pricing_raw and property_area are raw strings
# already parsed into price_per_aana and land_size_aana
# road_access already parsed into road_width_feet
# landmark is free text, not useful for ML
# date_posted not relevant for price prediction
drop_cols = ['pricing_raw', 'property_area', 'road_access', 
             'landmark', 'date_posted']
df_lp.drop(columns=drop_cols, inplace=True)

# ─────────────────────────────────────────
# FINAL VERIFICATION
# ─────────────────────────────────────────
print(f"\nFinal shape: {df_lp.shape}")
print(f"Nulls: {df_lp.isnull().sum().sum()}")
print(f"Columns: {list(df_lp.columns)}")
print(f"\nprice_per_aana range: {df_lp['price_per_aana'].min():,.0f} – {df_lp['price_per_aana'].max():,.0f}")
print(f"land_size_aana range: {df_lp['land_size_aana'].min()} – {df_lp['land_size_aana'].max()}")
```

```python
# ─────────────────────────────────────────
# COLUMN DESCRIPTIONS
# ─────────────────────────────────────────

# district          — Which of the 3 Kathmandu valley districts the property is in
#                     (Kathmandu, Lalitpur, Bhaktapur)

# property_type     — Type of land use classification
#                     (Residential, Commercial, Semi-commercial)

# property_face     — Which direction the plot faces
#                     (North, South, East, West, North-East etc.)

# road_type         — Quality of road access to the property
#                     (High Access = Blacktopped/Paved, Low Access = Gravelled/Soil/Alley)

# negotiable        — Whether the listed price is negotiable
#                     (Yes/No — useful signal for recommendation module)

# views             — Number of times the listing was viewed on lalpurja.com
#                     Higher views may indicate more desirable properties

# municipality      — The local government body the property falls under
#                     More granular than district (e.g. Budhanilkantha, Lalitpur)

# ward_no           — Ward number within the municipality
#                     Most granular location identifier available

# neighborhood      — Neighborhood name extracted from listing
#                     (e.g. Budhanilkantha, Imadol, Balkot)

# price_per_aana    — Parsed and converted price per aana in NPR
#                     Original was "35 Lakh Per Aana" → 3,500,000 NPR
#                     This is the TARGET variable for regression modeling

# land_size_aana    — Size of the plot in aana
#                     Parsed from "4 Aana", "5.5 Aana" etc.
#                     Capped at 50 aana to remove commercial anomalies

# road_width_feet   — Width of the road accessing the property in feet
#                     Parsed from road_access column e.g. "13 Feet" → 13

# facing_road_width — Width of the road the property directly faces in feet
#                     Parsed from dimension column e.g. "F 28 Feet" → 28
#                     Different from road_width — this is the frontage road

# hospital_m        — Distance to nearest hospital in meters
#                     Converted from "1km" → 1000, "500m" → 500
#                     Capped at 15,000m (15km) to remove parsing errors

# airport_m         — Distance to Tribhuvan International Airport in meters
#                     Capped at 30,000m (30km)

# pharmacy_m        — Distance to nearest pharmacy in meters
#                     Capped at 10,000m (10km)

# bhatbhateni_m     — Distance to nearest Bhatbhateni Supermarket in meters
#                     Proxy for urban amenity access and commercial development
#                     Capped at 15,000m (15km)

# school_m          — Distance to nearest school in meters
#                     Important for families — strong signal for recommendation
#                     Capped at 10,000m (10km)

# public_transport_m — Distance to nearest public transport stop in meters
#                     Bus stop, microbus stand etc.
#                     Capped at 5,000m (5km)

# police_station_m  — Distance to nearest police station in meters
#                     Proxy for safety and urban infrastructure
#                     Capped at 15,000m (15km)

# pashupatinath_m   — Distance to Pashupatinath Temple in meters
#                     Major landmark used as geographic reference point
#                     Capped at 30,000m (30km)

# boudhanath_m      — Distance to Boudhanath Stupa in meters
#                     Major landmark used as geographic reference point
#                     Capped at 30,000m (30km)

# ring_road_m       — Distance to Kathmandu Ring Road in meters
#                     Ring road proximity is a strong price signal in Kathmandu
#                     Properties near ring road command premium prices
#                     Capped at 30,000m (30km)

# ─────────────────────────────────────────
# SAVE
# ─────────────────────────────────────────
df_lp.to_csv('cleaned_lalpurja_land.csv', index=False)
print(f"✅ Saved cleaned_lalpurja_land.csv")
print(f"✅ Shape: {df_lp.shape}")
print(f"✅ Columns: {list(df_lp.columns)}")
```
